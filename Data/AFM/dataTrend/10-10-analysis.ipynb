```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
print(stif.shape)
```

```python
# reshape the dataframe for processing
stifvalue = stif.values.reshape(589056,1)
sample2 = pd.DataFrame(data = stifvalue, index = None)
print(sample2)
```
